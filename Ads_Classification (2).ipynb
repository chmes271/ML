```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('english')) 
def clean_text(text):
    clean_words = []
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text) 
    for word in tokens: 
        if word not in stop_words: 
            new_word = re.sub('[^a-zA-Z]', '',word)
            new_word = re.sub('r[,/\^_]','', new_word)
            new_word = new_word.lower()
            new_word = lemmatizer.lemmatize(new_word)
            clean_words.append(new_word)
 
    return ' '.join(clean_words)
```

```python
x1 = [clean_text(i) for i  in x]
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
cv = CountVectorizer(ngram_range=(1,2))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 1)
```

```python
x_train_vec = cv.fit_transform(x_train)
```

```python
x_test_vec = cv.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mn = MultinomialNB()
```

```python
mn.fit(x_train_vec, y_train)
```

```python
y_pred = mn.predict(x_test_vec)
```

```python
from sklearn.metrics import *
```

```python
accuracy = accuracy_score(y_pred, y_test)
```
