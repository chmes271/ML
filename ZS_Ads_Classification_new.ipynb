```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk import ne_chunk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.model_selection import train_test_split
```

```python
pst = PorterStemmer()
lmt = WordNetLemmatizer()
```

```python
stop_words = list(set(stopwords.words('english')))

def reg(text):
    tokens = word_tokenize(text)
    line1 =[]
    for word in tokens:
        if word not in stop_words:
            a = re.sub(r'[^a-zA-Z]', ' ', word)
            a = re.sub(r'[/\^_.]', '', a)
            a = re.sub('^[ \t]+|[ \t]+$', '', a)
            a = a.lower()
            line1.append(a)
    return ' '.join(line1)
```

```python
clean_text_1 = [reg(i) for i in x]
```

```python
clean_text_1
```

```python
def porter(text):
    line2 = []
    for word in text:
        b = pst.stem(word)
        line2.append(b)
    return ''.join(line2)
        
```

```python
clean_text_2 = [porter(i) for i in clean_text_1]
print(clean_text_2)
```

```python
def lemmat(text):
    line3 = []
    for word in text:
        c = lmt.lemmatize(word)
        line3.append(c)
    return ''.join(line3)
        
```

```python
clean_text_3 = [lemmat(i) for i in clean_text_1]
print(clean_text_3)
```

```python
x_train, x_test, y_train, y_test = train_test_split(clean_text_3,y, test_size = 0.2, random_state = 1)
```

```python
stop_words = set(stopwords.words('english')) 
def clean_text(text):
    clean_words = []
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text) 
    for word in tokens: 
        if word not in stop_words: 
            new_word = re.sub('[^a-zA-Z]', '',word)
            new_word = re.sub('r[,/\^_]','', new_word)
            new_word = new_word.lower()
            new_word = lemmatizer.lemmatize(new_word)
            clean_words.append(new_word)
 
    return ' '.join(clean_words)
```
